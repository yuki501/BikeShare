```python
def m_com_lat(i,j):
    for k in range(len(station_df)):
        if station_df.index[k] == morning_most_common_route.index[i][j]:
            return station_df.iloc[k][0]
            
def m_com_lon(i,j):
    for k in range(len(station_df)):
        if station_df.index[k] == morning_most_common_route.index[i][j]:
            return station_df.iloc[k][1]        
m_com_lon(0,1)
```

```python
from folium import plugins

m = folium.Map(location=[40.8, -74], 
               tiles='Stamen Toner',
               zoom_start=12)

attr = {'fill': 'red'}

for a in range(20):
    folium.Marker([m_com_lat(a,0), m_com_lon(a,0)], 
                  popup=morning_most_common_route.index[a][0]).add_to(m)
    folium.Marker([m_com_lat(a,1), m_com_lon(a,1)], 
                  popup=morning_most_common_route.index[a][1]).add_to(m)
    line = folium.PolyLine(
        [[m_com_lat(a,0), m_com_lon(a,0)],
         [m_com_lat(a,1), m_com_lon(a,1)],]
    ).add_to(m) 
    
    plugins.PolyLineTextPath(
        line,
        "\u25BA",
        repeat=True,
        offset=5,
        attributes=attr
    ).add_to(m)



    

m.save(os.path.join('results', 'morning_commonroute_citi_1702.html'))
m
```
